```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.options.display.max_columns = 15
pd.options.display.max_rows = 6
plt.style.use('seaborn-dark')

#https://www.ibge.gov.br/estatisticas/sociais/educacao/9221-sintese-de-indicadores-sociais.html?=&t=downloads
# Luara Cristine Lopes Goulart, Luiz Henrique da Silva
```

```python
filename1 = r'NiveldeInstrucaoBrasil.csv'
df = pd.read_csv(filename1, sep=";", error_bad_lines=False, encoding='latin-1')
```

```python
df
```

```python
reg = df['regioes'].values.tolist()
se = df['Sem ensino'].values.tolist()
cvse = df['CV SE'].values.tolist()
efin = df['EF incompleto'].values.tolist()
cvefin= df['CV EF Inc'].values.tolist()
efcom = df['EF completo'].values.tolist()
cvefcom= df['CV EF com'].values.tolist()
emin = df['EM incompleto'].values.tolist()
cvemin = df['CV EM inc'].values.tolist()
emcom = df['EM completo'].values.tolist()
cvemcom  = df['CV EM com'].values.tolist()
esin = df['ES incompleto'].values.tolist()
cvesin = df['CV ES inc'].values.tolist()
escom = df['ES completo'].values.tolist()
cvescom = df['CV ES com'].values.tolist()
```

```python
df = pd.DataFrame({'Sem instrução': se,'Ensino fundamental incompleto': efin,'Ensino fundamental completo': efcom,'Ensino médio incompleto': emin,'Ensino médio completo': emcom,'Ensino superior incompleto': esin,'Ensino superior completo ': escom }, index=reg)
ax = df.plot.bar(rot=0,figsize=(15,7))
ax.set_xlabel('Nível de instrução da população acima de 25 anos idade',fontsize=16)
ax.set_ylabel('Porcentagem',fontsize=16)
ax.set_ylim(0,50)
```

```python
f, ax = plt.subplots(figsize=(10, 4))

sns.set_color_codes("pastel")
sns.barplot(x=se, y=reg, data=df, label="Total por região", color="b")

sns.barplot(x=cvse, y=reg, data=df, label="Coeficiente de variação", color="r")

ax.legend(ncol=2, loc="upper center", frameon=True)
ax.set(xlim=(0, 24), ylabel="Regioes", xlabel="Porcentagem total da população Sem ensino ")
ax.set(ylim=(-1,6))
sns.despine(left=True, bottom=True, )
```

```python
f, ax = plt.subplots(figsize=(10, 4))

sns.set_color_codes("dark")
sns.barplot(x=efin, y=reg, data=df, label="Total por região", color="b")

sns.barplot(x=cvefin, y=reg, data=df, label="Coeficiente de variação", color="c")

ax.legend(ncol=2, loc="upper center", frameon=True)
ax.set(xlim=(0, 42), ylabel="Regioes", xlabel="Porcentagem total da população com \n ensino fundamental incompleto")
ax.set(ylim=(-1,6))
sns.despine(left=True, bottom=True, )
```

```python
f, ax = plt.subplots(figsize=(10, 4))

sns.set_color_codes("deep")
sns.barplot(x=emin, y=reg, data=df, label="Total por região", color="y")

sns.barplot(x=cvemin, y=reg, data=df, label="Coeficiente de variação", color="g")

ax.legend(ncol=2, loc="upper center", frameon=True)
ax.set(xlim=(0, 6), ylabel="Regioes", xlabel="Porcentagem total da população com \n ensino medio incompleto")
ax.set(ylim=(-1,6))
sns.despine(left=True, bottom=True, )
```

```python
f, ax = plt.subplots(figsize=(10, 4))

sns.set_color_codes("deep")
sns.barplot(x=esin, y=reg, data=df, label="Total por região", color="c")

sns.barplot(x=cvesin, y=reg, data=df, label="Coeficiente de variação", color="r")

ax.legend(ncol=2, loc="upper center", frameon=True)
ax.set(xlim=(0, 6), ylabel="Regioes", xlabel="Porcentagem total da população com \n ensino superior incompleto")
ax.set(ylim=(-1,6))
sns.despine(left=True, bottom=True, )
```
